# Importing Libraries


```python
##Scipy functions
import scipy.io.wavfile as wv
from scipy.fftpack import fft

##Others
import numpy as np
import os
import matplotlib.pyplot as plt
```

# Functions


## Read .wav files as signals


```python
DATA_FOLDER_PATH = "../Data"
FILE = "string_1.wav"


filepath = os.path.join(DATA_FOLDER_PATH, FILE)
```

```python
sampleFreq, audio = wv.read(filepath)
sampleLen = len(audio)/sampleFreq
```

```python
timeX = np.arange(0, sampleLen, 1/sampleFreq)

plt.plot(timeX, audio)
plt.xlabel("Time (s)")
plt.ylabel("Signal Magnitude (s(t))")
plt.title(f"Audio File : {FILE}")

plt.show()
```

## FFT of Signal


```python
freqSpectrum = abs(fft(audio))
```

```python
freqX = np.arange(0, sampleFreq/2, 1/sampleLen)

plt.plot(freqX, freqSpectrum[:len(freqX)])


plt.xlabel("Frequency (Hz)")
plt.ylabel("Frequency Magnitude (|X(t)|)")
plt.title(f"Audio File : {FILE}")

plt.show()
```

## Find maximum Frequency


```python
np.argmax(freqSpectrum) / sampleLen
```
